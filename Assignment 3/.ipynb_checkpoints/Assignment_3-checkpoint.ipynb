# L-2 regularized SVM Optimization Problem

The loss function is - $$
f(\mathbf{w}, b)=\frac{1}{n} \sum_{i=1}^n\left(1-y_i\left(\left\langle\mathbf{w}, \mathbf{x}_i\right\rangle+b\right)\right)_{+}+\frac{\lambda}{2}\|\mathbf{w}\|_2^2
$$
Here $\mathbf{x}_i \in \mathbb{R}^d$ are feature vectors and $y_i \in\{-1,1\}$.

### Q: What is the subdifferential of $f$?

The subdifferential of $f$ can be computed as -
$$ 
\begin{align}
\partial_{\mathbf{\omega}} f 
& =\; \frac{1}{n} \sum_{i=1}^n \partial_{\mathbf{\omega}} \left(1-y_i\left(\left\langle\mathbf{w}, \mathbf{x}_i\right\rangle+b\right)\right)_{+} + \frac{\lambda}{2} \partial_{\mathbf{\omega}} \left(\|\mathbf{w}\|_2^2 \right) \\
\end{align}
$$

Calculating the derivative w.r.t $\mathbf{w}$-
$$
\begin{align}
\partial_{\mathbf{w}} \left(1-y_i\left(\left\langle\mathbf{w}, \mathbf{x}_i\right\rangle+b\right)\right)_{+}
& =\; - y_i x_i                  \; &&  y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) < 1  \\
& =\; 0                          \; &&  y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) > 1  \\
& =\; \left[-y_i x_i, 0 \right]  \; &&  y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) = 1. \\
\end{align}
$$

$$
\partial_{\mathbf{w}} \left(\|\mathbf{w}\|_2^2 \right) = 2 \mathbf{w}
$$


Calculating the derivative w.r.t. $b$-


$$
\begin{align}
\partial_{b} \left(1-y_i\left(\left\langle\mathbf{w}, \mathbf{x}_i\right\rangle+b\right)\right)_{+}
& =\; - y_i                  \; && y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) < 1  \\
& =\; 0                      \; && y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) > 1  \\
& =\; \left[-y_i, 0 \right]  \; && y_i\left(\mathbf{w}^T \mathbf{x}_i + b\right) = 1. \\
\end{align}
$$

$$
\partial_{b} \left(\|\mathbf{w}\|_2^2 \right) = 0
$$


```python
import numpy as np
from joblib import Memory
from sklearn.datasets import load_svmlight_file
from sklearn.preprocessing import StandardScaler
from sklearn.utils.random import sample_without_replacement
```

```python
def sgdmethod(X, Y, subgradloss, subgradreg, regparam, w1, T, a, m):
    out = []
    w,b = w1
    out.append((w,b))
    subgradloss_w = np.zeros([X.shape[1]])
    subgradloss_b = 0
    loss = 0
    objfun = np.zeros(T)
    p,n = X.shape
    for t in range(0, T):
        sample_index = sample_without_replacement(p,m) ## Sampling the random index from minibatch 
        learning_rate = 1/(1 + a * t)                  ## Defining the learning rate
        x = X[sample_index,:] 
        y = Y[sample_index]
        for i in range(m):
            loss_i, sub_w , sub_b = subgradloss(x[i,:], y[i], (w,b))  ## calculating subgrad for a single sample given parameters w
            loss += loss_i
            subgradloss_w += sub_w
            subgradloss_b += sub_b
        reg, subgrad_w, subgrad_b = subgradreg(regparam, (w,b)) ## Adding the regularizer term 
        objfun[t] = hingelossobj(x, y, (w,b), regparam)
        w = w - learning_rate * (subgradloss_w / m + subgrad_w)
        b = b - learning_rate * (subgrad_b /m + subgrad_b)
        out.append((w,b))
    return(out,objfun)

def subgradloss(x, y, w):
    W,b = w
    f_x = np.dot(x,W) + b
    loss_val = 1 - y * f_x
    if loss_val >= 0:
        subgrad_w = - y * x
        subgrad_b = -1 * y
    else:
        loss_val = 0 
        subgrad_w = 0 * x
        subgrad_b = 0
    return((loss_val, subgrad_w, subgrad_b))

def subgradreg(regparam, w):
    W,b = w
    reg = regparam / 2 * np.dot(W,W)
    subgrad_w_reg = regparam * W
    subgrad_b_reg = 0
    return(reg,  subgrad_w_reg, subgrad_b_reg)

def hingelossobj(x, y, w, lam):
    W, b = w
    b = np.ones(x.shape[0])*b
    dis = 1 - y * (np.dot(x, W) + b)
    dis[dis < 0] = 0 
    objfun = lam/2 * np.dot(W, W) + np.sum(dis) / x.shape[0]
    return objfun
```

```python
mem = Memory("./mycache")

@mem.cache
def get_data(file_name):
    data = load_svmlight_file(file_name)
    return data[0], data[1]

X, y = get_data("a9a.txt")
X_test, y_test = get_data("a9a_t.txt")
X = X[:,:122].toarray()
X_test = X_test.toarray()
```

```python
n_train = X.shape[0]
regparam = 1/n_train
w = np.ones((X.shape[1]))
b = 1
w1 = (w,b)
T = 1000
```

### Evaluation of the learned model 

The accuracy metric is used for evaluation of the model. 

```python
def accuracy(X, w, T, y):
    result = np.zeros(T)
    for t in range(T):
        y_pred = np.sign(np.dot(X,w[t][0]) + w[t][1] * np.ones(X.shape[0]))
        score = np.where(y_pred == y)
        result[t] = len(score[0]) / X.shape[0] * 100
    return(result)
```

### Last half averaging 

```python
def wei_bi(w):
    wei = []
    bi = []
    for t in range(T):
        wei.append(w[t][0])
        bi.append(w[t][1])
    return(wei,bi)
```

```python
def averaging(wei,bi, X,y, T):
    last_half_accuracy = np.zeros(T)
    weight = wei[:]
    bias = bi[:]
    for t in range(2,T):
        weight[t] = (1/(t - np.floor(t/2)+1)) * sum(wei[int(np.floor(t/2)):t]) 
        bias[t] = (1/(t - np.floor(t/2)+1)) * sum(bi[int(np.floor(t/2)):t])
        y_pred = np.sign(np.dot(X,weight[t]) + bias[t] * np.ones(X.shape[0]))
        score = np.where(y_pred == y)
        last_half_accuracy[t] = len(score[0]) / X.shape[0] * 100
    return(last_half_accuracy)
```

### Accuracy on Training Data

```python
wei, bi = wei_bi(w_full)
last_half_accuracy_train = averaging(wei, bi, X,y, T)
```

```python
plt.plot(last_half_accuracy_train, linewidth = '2')
plt.title('accuracy')
plt.xlabel('Iterations'); plt.ylabel('Accuracy (%)')
plt.show()
```

## Comparison between Stochastic Gradient Descent and Sub-gradient Descent Methods

- Gradient Descent (GD) is a batch optimization algorithm that updates the model parameters using the gradient of the loss function computed over the entire training set, while Stoshastic Gradient Descent (SGD) updates the parameters using the gradient computed over a randomly selected subset (mini-batch) of the training set at each iteration.

- SGD typically converges faster than GD, especially for large datasets. This is because GD computes the gradient over the entire dataset, which can be computationally expensive and time-consuming. On the other hand, SGD uses a mini-batch, which reduces the computational cost and allows for faster updates to the model parameters.

- SGD is prone to oscillations and can get stuck in local minima, whereas GD is more stable and can converge to the global minimum. 

- The averaging methods computes an average of the gradients over previous iterations. This averaging helps to reduce the variance and oscillations of the updates and can lead to better convergence. Which is evident from the results too.

## Part - 1

The following expression represents an update rule for the optimization variable $\mathbf{x}_{t+1}$, where $C$ is a convex set, $f$ and $g$ are convex functions of $\mathbf{x}_t$:


$$
\begin{align}
\mathbf{x}_{t+1} & = \text{argmin}_{\mathbf{u} \in C} f(\mathbf{x}_t) + g(\mathbf{x}_t)^T (\mathbf{u} - \mathbf{x}_t) +                              \cfrac{1}{2 \alpha_t} \| \mathbf{u} - \mathbf{x}_t \|_2^2 \\
                 & = \text{argmin}_{\mathbf{u}} \cfrac{1}{2 \alpha_t} \left(  2 \alpha_t g(\mathbf{x}_t)^T \mathbf{u} +                              \|\mathbf{u} \|_2^2 - 2 \mathbf{x}_t^T \mathbf{u} \right) \\
\end{align}
$$

This expression can be simplified by completing the square and rearranging terms:

$$
\begin{aligned}
\mathbf{x}_{t+1} & =\operatorname{argmin}_{\mathbf{u}} \frac{1}{2 \alpha_t}\left\|\mathbf{u}-\left(\mathbf{x}_t-\alpha_t g\left(\mathbf{x}_t\right)\right)\right\|_2^2 \\
& =\operatorname{argmin}_{\mathbf{u}} P_C\left(\mathbf{x}_t-\alpha_t g\left(\mathbf{x}_t\right)\right)
\end{aligned}
$$


Here, $P_C$ represents a projection function that projects the vector $\mathbf{x}_t - \alpha_t g(\mathbf{x}_t)$ onto the convex set $C$. This update rule is a convex optimization problem, and the projection ensures that the solution remains within the convex set.

Geometrically, this update rule projects the objective onto the convex set $C$ to ensure that the solution remains bounded and does not blow up over time. The third term in the update expression minimizes the distance between the current and next iterate, ensuring that the solution remains within the convex set.

## Part - 2

To improve the expression for $\mathbf{x}_{t+1}$, we can rearrange the terms in the equation by neglecting the terms that are independent of $\mathbf{x}$:

\begin{aligned}
\mathbf{x}_{t+1} & =\operatorname{argmin}_{\mathbf{u}} f\left(\mathbf{x}_t\right)+\mathbf{g}^T\left(\mathbf{u}-\mathbf{x}_t\right)+\frac{1}{2 \alpha_t}\left\|\mathbf{P}_t^{1 / 2}\left(\mathbf{u}-\mathbf{x}_t\right)\right\|_2^2 \\
& =\operatorname{argmin}_{\mathbf{u}} \frac{1}{2 \alpha_t}\left(2 \alpha_t \mathbf{g}^T \mathbf{u}+\mathbf{P}_t\left(\|\mathbf{u}\|_2^2-2 \mathbf{x}_t^T \mathbf{u}\right)\right) \\
& =\operatorname{argmin}_{\mathbf{u}} \frac{1}{2 \alpha_t} \mathbf{P}_t\left(\| \mathbf{u}-\left(\mathbf{x}_t-\alpha_t \mathbf{P}_t^{-1} \mathbf{g} \|_2^2\right)\right. \\
& =\operatorname{argmin}_{\mathbf{u}} P_C\left(\mathbf{x}_t-\alpha_t \mathbf{P}_t^{-1} \mathbf{g}\right)
\end{aligned}

Here, $P_C$ is a valid projection on a convex set because the Hessian $\mathbf{P}_t$ is an affine conversion on the 2-norm ball and satisfies $\mathbf{P}_t \geq 0$.

The projected Newton's algorithm is the same as the differentiable case ($\mathbf{g} = \nabla f(\mathbf{x})$):

$$\mathbf{x}_{t+1}=P_C\left(\mathbf{x}_t-\mathbf{P}_t^{-1} \nabla f\left(\mathbf{x}_t\right)\right)$$

In the case of AdaGrad, an approximation to the Hessian is used instead of the exact Hessian. When the approximation has minimum error, the Adagrad scheme iterates will be equal to the present iterates.

# Question - 3

We make the assumption that the error between the exact Newton and Newton-CG solutions can be approximated by the error between the minimum of the quadratic model and the gradient of the function at the current point, i.e., $\text{error}(f(x)_{\text{exact Newton}} - f(x)_{\text{Newton-CG}} \approx \text{error}(\text{argmin}_x || \nabla^2 f(x_t) x - \nabla f(x_t)||_2^2)$.

Using this assumption, we can bound the error of the Newton-CG method as follows:

By setting $z_1$ to 0, we can obtain an upper bound on the error as $\left(1-\frac{1}{\sqrt{k}}\right)^{T-1}||\nabla f(x_t)||_2^2$. 

$$
\begin{align}
error & \leq \left( 1 - \cfrac{1}{\sqrt{k}} \right)^{T-1} B \quad \text{using the bound on $\nabla f(x)$} \\ 
      & \leq \left( 1 - \cfrac{1}{\sqrt{O \left( d^2 ln \left( \cfrac{B}{\epsilon} \right)^{-2} \right) }} \right)^{T-1} B \\ 
      & \leq \left( 1 - \cfrac{1}{O \left( d\; ln \left( \cfrac{B}{\epsilon} \right)^{-1} \right) } \right)^{T-1} B \\ 
      & \leq \left( 1 - O \left( \cfrac{1}{d}\; \cfrac{1}{ln \left( \cfrac{\epsilon}{B} \right)} \right) \right)^{T-1} B \\ 
      & \leq \left( 1 - O \left( \cfrac{1}{d}\; \cfrac{1}{\cfrac{\epsilon}{B}} -1 \right) \right)^{T-1} B \quad \text{using the inequality $ ln(x) \leq x - 1\; \text{when}\; x > 0$} \\
      & \leq O \left( \cfrac{d (\epsilon - B) - B}{d (\epsilon - B)} \right)^{T-1} B \\
      & \leq O \left( \cfrac{d (\epsilon - B) - B}{d (B)} \right)^{T-1} B \quad \text{since $\epsilon << B$} \\
      & \leq O \left( \cfrac{d (\epsilon_2 - B) - B}{d} \right) \quad \text{error for T=2, i.e., evaluating $z_2$ moving from $z_1$ = 0} \\
      & \leq O (\epsilon_2 - B) \quad \text{since B << d} \\
\end{align}
$$

This is better than the error bound of $O(\epsilon)$ for the steepest descent method used in the exact Newton method.
