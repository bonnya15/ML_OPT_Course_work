### Let $\sigma(x)=x_{+}$denote the ReLU activation function. Write the function $s$ given below using a one hidden layer ReLU neural network containing 2 neurons.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
	return(max(0.0, x))


def s(x):
    return(relu(x + 1) - relu(x))

x = np.arange(-2,1,0.2)

y = np.zeros(x.shape).reshape(-1)

for i in range(len(x)):
    y[i] = s(x[i])
    
plt.plot(x,y)
plt.show()
```

### Express the spike $f$ function below in terms of $s$, and give a two-hidden layer ReLU neural network containing 5 neurons that computes $f$.

```python
def both(x):
    s1 = s(x)
    s2 = s(-x)
    return(relu(s1 + s2 - 1))


x = np.arange(-2,2,0.2)

y = np.zeros(x.shape).reshape(-1)

for i in range(len(x)):
    y[i] = both(x[i])
    
plt.plot(x,y)
plt.show()
```
