# Question - 1
### Soft-shrinkage operator

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def softShrink(x0, alpha):
    x0 = x0.reshape(len(x0),1)
    x = np.zeros((len(x0),1))
    x[x0 > alpha] = x0[x0 > alpha] - alpha
    x[x0 < -alpha] = x0[x0 < -alpha] + alpha
    return(x)
```

```python
x0 = np.arange(-10, 10, 1, dtype=int)
alpha = 3
```

```python
x = softShrink(x0, alpha)
```

```python
plt.plot(x0, x)
plt.show()
```

```python
def istaLasso(A, b, lambda_reg, x0, alpha, T):
    objhist = [f(A,x0,b)+g(lambda_reg,x0)]
    X = [x0]
    for t in range(T):
        x0_softShrink = x0 - alpha*subgrad(A,x0,b)
        alpha_softShrink = alpha*lambda_reg
        x = softShrink(x0_softShrink, alpha_softShrink).reshape(-1)
        X.append(x)
        objhist.append(f(A,x0,b)+g(lambda_reg,x0))
        x0 = x
    xT = X[-1]
    return(xT, objhist)

def f(A,x,b):
    return(0.5 * np.linalg.norm(A.dot(x) - b)** 2)
def g(lambda_reg, x):
    return(lambda_reg * np.linalg.norm(x,1))
def subgrad(A,x,b):
    ols_term = np.dot(A.T, A.dot(x) - b)
    return(ols_term)
```

```python
m = 15
n = 20
T = 3000
lambda_reg = 0.5
alpha = 0.001
A = np.random.randn(m, n)  # random design
xo = np.random.rand(n)     
xo[xo < 0.9] = 0     # Final Solution
b = np.dot(A, xo)
x0 = np.zeros(A.shape[1]) ## Initial solution

xT, objhist = istaLasso(A, b, lambda_reg, x0, alpha, T)
```

```python
xT
```

```python
xo
```

```python
plt.plot(range(len(objhist)), objhist)
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```

# Question - 4

```python
def lassoSubgrad(A, b, lambda_reg, x):
    ols_term = np.dot(A.T, A.dot(x) - b)
    lasso_term = lambda_reg * np.sign(x)
    return(ols_term + lasso_term)

def subgradLasso(A, b, lambda_reg, x0, alpha, T):
    objhist = [f(A,x0,b)+g(lambda_reg,x0)]
    X = [x0]
    for t in range(T):
        x = x0 - alpha*lassoSubgrad(A,b, lambda_reg, x0)
        X.append(x)
        objhist.append(f(A,x0,b)+g(lambda_reg,x0))
        x0 = x
    xT = X[-1]
    return(xT, objhist)

def f(A,x,b):
    return(0.5 * np.linalg.norm(A.dot(x) - b)** 2)
def g(lambda_reg, x):
    return(lambda_reg * np.linalg.norm(x,1))
```

```python
m = 15
n = 20
T = 3000
lambda_reg = 0.5
alpha = 0.001
A = np.random.randn(m, n)  # random design
xo = np.random.rand(n)     
xo[xo < 0.9] = 0     # Final Solution
b = np.dot(A, xo)
x0 = np.zeros(A.shape[1]) ## Initial solution

xT, objhist = subgradLasso(A, b, lambda_reg, x0, alpha, T)
```

```python
xT
```

```python
xo
```

```python
plt.plot(range(len(objhist)), objhist)
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```

# Question - 5

```python
from deblurringUtils import * 
from skimage import data
import numpy.random as random
```

```python
n = 128 
camera = data.camera() 
smallcamera = resize(camera, [n, n]) 
camdata = img_as_float(smallcamera) 
visualize(camdata)
```

```python
H = vectorized2DHaarMatrix(n, n)            # the Haar matrix, an orthonormal basis 
haarcoeffs = H.T @ vectorize(camdata)       # convert to the Haar basis 
p = 90                                      # what percentage of the coefficients to set to zero 
thres = np.percentile(np.abs(haarcoeffs), p) 
haarcoeffs[abs(haarcoeffs) <= thres] = 0    # set most entries to zero 
# reconstruct the image from the sparse representation 
reconstim = unvectorize(H @ haarcoeffs, n, n) 
visualize(reconstim)
```

```python
# mask the same percentage of pixels, randomly 
mask = np.ones(n**2) 
mask[:int(p/100*n**2)] = 0 
mask = unvectorize(random.permutation(mask), n, n) 
visualize(camdata * mask )
```

```python
B = vectorized2DBlurMatrix(n, n, 5); 
std = 1e-2 
corruption = std * random.randn(n**2) 
b = B @ vectorize(smallcamera) + corruption 
blurredcam = unvectorize(b, n, n) 
visualize(blurredcam)
```

### De-blurring using naive method

```python
from scipy.sparse.linalg import gmres 
from scipy.sparse import csr_matrix

# solving with sparse matrices is faster 
sB = csr_matrix(B) 
linres, _ = gmres(sB, b, maxiter=50) 
visualize(unvectorize(linres, n, n))
```

```python
linres
```

### Step size of the subgradient method

```python
from scipy.sparse.linalg import svds 
A = B @ H 
_, topsv, _ = svds(A, k=1) 
alphaSubgrad = 1/topsv**2
```

```python
print("The step size of the subgradient method is : {}".format(alphaSubgrad))
```

### De-blurring with Haar matrix 

```python
lambda_reg = 2*(10**(-4))
T = 1000
x0 = np.zeros(b.shape[0])
```

#### (a) Using subgradLasso 

```python
xT, objhist = subgradLasso(A, b, lambda_reg, x0, alphaSubgrad, T)
```

```python
print("The final objective value is : {}".format(objhist[-1]))
```

```python
visualize(unvectorize(H @ xT, n, n))
```

```python
plt.plot(range(len(objhist)), objhist)
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```

#### (b) Using istaLasso

```python
xT_lasso, objhist_lasso = istaLasso(A, b, lambda_reg, x0, alphaSubgrad, T)
```

```python
print("The final objective value is : {}".format(objhist_lasso[-1]))
```

```python
visualize(unvectorize(H @ xT_lasso, n, n))
```

```python
plt.plot(range(len(objhist_lasso)), np.log(objhist_lasso))
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```
