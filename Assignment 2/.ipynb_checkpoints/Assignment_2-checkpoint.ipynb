# Question - 1
### Soft-shrinkage operator

There is a unique minimizer to the optimization problem,
$$ S_{\alpha} (\mathbf{x_0}) = \text{argmin}_{\mathbf{x} \in \mathbb{R}^d} \cfrac{1}{2 \alpha} \| \mathbf{x} - \mathbf{x}_0 \|_2^2 + \| \mathbf{x} \|_1 $$

1. $ \mathbf{x} - \mathbf{x}_0 $ is an affine function for $ \mathbf{x}_0 \in \mathbf{R}^d $.
2. Square of any function is strictly convex.
3. Multiplication with non negative constant is also convex (here, $\alpha$ > 0).
4. 1-norm of $ \mathbf{x} $ is an affine function and we know that any affine function is convex function. 
5. The domain of the objective function is $\mathbb{R}^d$ is a convex set.

As, $\cfrac{1}{2 \alpha} \| \mathbf{x} - \mathbf{x}_0 \|_2^2$ is strictly convex, adding this term with other convex function makes the objective strictly convex. We also know that, strictly convex function has a unique minimizer.

For the function, $ f = \cfrac{1}{2 \alpha} \| \mathbf{x} - \mathbf{x}_0 \|_2^2 + \| \mathbf{x} \|_1$, the set of subgradients is defined as,


$ g =  \begin{cases} \cfrac{1}{\alpha} (x_i - (x_0)_i) + 1  & \text{for}~ x_i > 0 \\
     \cfrac{1}{\alpha} (x_i - (x_0)_i) - 1  & \text{for}~ x_i < 0 \\
     \left[ \left(-\cfrac{(x_0)_i}{\alpha}-1\right), \left(-\cfrac{(x_0)_i}{\alpha}+1 \right)\right]   & \text{for} ~ x_i = 0
     \end{cases}
$

And the soft-shrinkage operator is defined as,



$s_{\alpha}((x_0)_i) = \begin{cases} (x_0)_i - \alpha  & \text{for}~ (x_0)_i > \alpha \\
                        (x_0)_i + \alpha  & \text{for}~ (x_0)_i < -\alpha \\
                        0  & \text{for} -\alpha \leq (x_0)_i \leq \alpha
                        \end{cases} $

According to the Fermat's optimality condition, i.e., for $g = 0$, the unique minimizer of $x_i$ is, 
$ S_{\alpha} (x_0)_i = s_{\alpha} ((x_0)_i) $ which is the soft-shrinkage operator.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def softShrink(x0, alpha):
    x0 = x0.reshape(len(x0),1)
    x = np.zeros((len(x0),1))
    x[x0 > alpha] = x0[x0 > alpha] - alpha
    x[x0 < -alpha] = x0[x0 < -alpha] + alpha
    return(x)
```

```python
x0 = np.arange(-10, 10, 1, dtype=int)
alpha = 3
```

```python
x = softShrink(x0, alpha)
```

```python
plt.plot(x0, x)
plt.show()
```

# Question - 2

Proximal gradient descent (aka composite gradient descent, generalized gradient descent, or the prox-linear algorithm) is used to solve problems of the form
$$
\min _{\mathbf{x}} f(\mathbf{x})+g(\mathbf{x})
$$
where $f$ is a differentiable convex function and $g$ is a convex function. To move from iterate $\mathbf{x}_t$ to $\mathbf{x}_{t+1}$, the method takes a gradient descent step of $f$, then chooses $\mathbf{x}_{t+1}$ to balance between minimizing $g(\mathbf{x})$ and staying close to the intermediate iterate:
$$
\mathbf{x}_{t+1}=\operatorname{argmin}_{\mathbf{x}} \frac{1}{2 \alpha_t}\left\|\mathbf{x}-\left(\mathbf{x}_t-\alpha_t \nabla f\left(\mathbf{x}_t\right)\right)\right\|^2+g(\mathbf{x}) .
$$
Note that (Prox) is always strictly convex, so $\mathbf{x}_{t+1}$ is always a well-defined single point. Argue that if $\mathbf{x}^{\star}$ is a fixed point of (Prox) then it is a minimizer of (Comp).
The proximal gradient algorithm is frequently employed when $g$ is nonsmooth. A canonical example of this use case is the ISTA algorithm below for the LASSO problem.

### Answer - 

According to Fermat's optimality condititon for unconstrained optimization, if $x^* \in \operatorname{argmin}_{x} f(x)$ iff $ 0 \in \partial f(x^*) $

I will show below that if 
$$
\mathbf{x^*}= \operatorname{prox}_{\alpha_t}\left[x^*-\alpha_t \nabla f\left(x^*\right)\right] = \operatorname{argmin}_{\mathbf{x}} \frac{1}{2 \alpha_t}\left\|\mathbf{x}-\left(\mathbf{x^*}-\alpha_t \nabla f\left(\mathbf{x^*}\right)\right)\right\|^2+g(\mathbf{x}) .
$$


Now since $x^*=\operatorname{prox}_{\alpha_t}\left[x^*-\alpha_t \nabla f\left(x^*\right)\right]$ is the fixed point in the proxy gradient descent, we have by Fermat's rule at $x=x^*$, the following
$$
0 \in \partial(g(x))+\frac{1}{\alpha_t}\left(x-\left(x^*-\alpha_t \nabla f\left(x^*\right)\right)\right)
$$
Now just substitute $x=x^*$, we get
$$
0 \in \alpha_t \partial g\left(x^*\right)+\alpha_t \nabla f\left(x^*\right)
$$
$$
0 \in \partial g\left(x^*\right)+ \nabla f\left(x^*\right)
$$
This is equivalent to saying that $0 \in \partial F\left(x^*\right)$, where $F(x)=g(x)+f(x)$, so $x^*$ is the minimizer. 

# Question - 3

For the LASSO optimization problem,
$$ \mathbf{x}^* = \text{argmin}_{\mathbf{x} \in \mathbb{R}^d} \cfrac{1}{2} \| \mathbf{Ax} - \mathbf{b} \|_2^2 + \lambda \|\mathbf{x}\|_1 $$
The composit gradient descent method can be written as,
$$
\begin{align}
\mathbf{x}_{t+1} & = \text{argmin}_{\mathbf{x}} \cfrac{1}{2 \alpha_t} \|\mathbf{x} - \left( \mathbf{x}_t - \alpha_t \mathbf{A}^T (\mathbf{Ax} - \mathbf{b}) \right) \|_2^2 + \lambda \| \mathbf{x} \|_1 
\end{align}
$$
This can be solved using the soft-shrinkage operator.

```python
def istaLasso(A, b, lambda_reg, x0, alpha, T):
    objhist = [f(A,x0,b)+g(lambda_reg,x0)]
    X = [x0]
    for t in range(T):
        x0_softShrink = x0 - alpha*subgrad(A,x0,b)
        alpha_softShrink = alpha*lambda_reg
        x = softShrink(x0_softShrink, alpha_softShrink).reshape(-1)
        X.append(x)
        objhist.append(f(A,x0,b)+g(lambda_reg,x0))
        x0 = x
    xT = X[-1]
    return(xT, objhist)

def f(A,x,b):
    return(0.5 * np.linalg.norm(A.dot(x) - b)** 2)
def g(lambda_reg, x):
    return(lambda_reg * np.linalg.norm(x,1))
def subgrad(A,x,b):
    ols_term = np.dot(A.T, A.dot(x) - b)
    return(ols_term)
```

```python
m = 15
n = 20
T = 3000
lambda_reg = 0.5
alpha = 0.001
A = np.random.randn(m, n)  # random design
xo = np.random.rand(n)     
xo[xo < 0.9] = 0     # Final Solution
b = np.dot(A, xo)
x0 = np.zeros(A.shape[1]) ## Initial solution

xT, objhist = istaLasso(A, b, lambda_reg, x0, alpha, T)
```

```python
xT
```

```python
xo
```

```python
plt.plot(range(len(objhist)), objhist)
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```

# Question - 4

For the LASSO objective function, 
$$ f = \cfrac{1}{2} \| \mathbf{Ax} - \mathbf{b} \|_2^2 + \lambda \|\mathbf{x}\|_1 $$
the subdifferential can be defined as,

$\partial f =  \begin{cases} \mathbf{A}^T (\mathbf{Ax} - \mathbf{b}) + \lambda & \text{for}~ \mathbf{x} > 0 \\
                \mathbf{A}^T (\mathbf{Ax} - \mathbf{b}) - \lambda \; & \text{for}~ \mathbf{x} < 0 \\
                \left[ \left( -\mathbf{A}^T \mathbf{b} - \lambda \right), \left(-\mathbf{A}^T \mathbf{b} + \lambda                               \right)\right]   & \text{for}~ \mathbf{x} = 0  
                \end{cases}
$

Here, we have solved the problem using subgradient descent method. 

```python
def lassoSubgrad(A, b, lambda_reg, x):
    ols_term = np.dot(A.T, A.dot(x) - b)
    lasso_term = lambda_reg * np.sign(x)
    return(ols_term + lasso_term)

def subgradLasso(A, b, lambda_reg, x0, alpha, T):
    objhist = [f(A,x0,b)+g(lambda_reg,x0)]
    X = [x0]
    for t in range(T):
        x = x0 - alpha*lassoSubgrad(A,b, lambda_reg, x0)
        X.append(x)
        objhist.append(f(A,x0,b)+g(lambda_reg,x0))
        x0 = x
    xT = X[-1]
    return(xT, objhist)

def f(A,x,b):
    return(0.5 * np.linalg.norm(A.dot(x) - b)** 2)
def g(lambda_reg, x):
    return(lambda_reg * np.linalg.norm(x,1))
```

```python
m = 15
n = 20
T = 3000
lambda_reg = 0.5
alpha = 0.001
A = np.random.randn(m, n)  # random design
xo = np.random.rand(n)     
xo[xo < 0.9] = 0     # Final Solution
b = np.dot(A, xo)
x0 = np.zeros(A.shape[1]) ## Initial solution

xT, objhist = subgradLasso(A, b, lambda_reg, x0, alpha, T)
```

```python
plt.plot(range(len(objhist)), objhist)
plt.xlabel("Iterations")
plt.ylabel("Loss function")
plt.title("Evolution of Objective Function through iterations")
plt.show()
```

### Merits of solving Lasso problem using subgradient and ista solvers 

In this example, 
- The performance of both the optimization methods are close enough. 
- The non-differentiable convex part of the $Prox(x)$, is relatively simpler to evaluate than the differentiable part.
- ISTA can be faster than the subgradient method. 
- But sometimes, choosing a suitable proxy objective function for solving a gradient descent algorithm might be challenging. In this case, it converged well. But in some other neumerical example choosing wrong proxy function might affect the convergence.  
- On the other hand, subgradient can be defined on a wide range of functions, which can be very easy to apply. 

### Advantages and disadvantages of solving the LASSO problem vs the least squares problem

The Lasso problem considers the sparse representation of the image, which essentially encodes the spatial properties of the image. But in case of only least square regression optimization, It is basically reconstructing the image by just looking into the average loss. So, no spatial information is preserved and as the background pixels are more dominant, the image gets almost dissolved in the background, even if the loss function is getting optimized. 

But in the sparse representation, we mask the background pixels in the sparse matrix and the darker pixels have some meaningful spatial representation. That is why, when we use Lasso it enforces sparcity of the features. Which somehow tries to preserve the structure of the image. 

That is why istaLasso and subgradLasso is performing better than the LS estimates. 

### Alternate Metric

The metric which is used for computing the quality of the solution is 
$$ \mathbf{z}^* = \text{argmin}_{\mathbf{z} \in \mathbb{R}^d} \cfrac{1}{2} \| \mathbf{Az} - \mathbf{b} \|_2^2 + \lambda \|\mathbf{z}\|_1 $$ It is basically considering all the pixels together. If we can break down the entire image in certain kernel size and compute the loss function for each kernel, the metric will be more meaningful. 

The kernels at different location will provide us the spatial information for the image. For smaller kernels (say 5 x 5) the estimation will be better. This is the technique used for convolutional neural networks where, kernels slide through the entire image for finding spatial features.
